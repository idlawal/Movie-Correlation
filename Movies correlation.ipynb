```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
movie = pd.read_csv('movies.csv')
movie.head(3)
```

```python
movie.tail()
```

```python
movie.describe()
```

```python
movie.dtypes
```

```python
#checking for missing values
for col in movie.columns:
    pct_missing = np.mean(movie[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
movie.isna().any()
```

```python
movie.isna().sum()
```

```python
movie.isna().sum().plot(kind='bar')
plt.show()
```

```python
movie.dropna(subset=['budget', 'gross'], inplace = True)
movie.head()
```

```python
#dropping duplicate
movie.drop_duplicates()
movie.head(3)
```

```python
# convert some columns into data type of Intger 'int64' 
movie['budget'] = movie['budget'].astype('int64')
movie['gross'] = movie['gross'].astype('int64')
movie.head()
```

```python
#Extracting column real_year from released column
movie['real_year'] = movie['released'].str.extract(pat = '([0-9]{4})').astype(int)
print(movie['real_year'].head())

```

```python
movie.head(2)
```

```python
#ploting a scatter plot
fig = plt.figure(figsize = (12,6))
sns.regplot(x='budget', y='gross', data=movie, scatter_kws={'color':'red'}, line_kws={'color': 'green'})
plt.scatter(x= movie['budget'], y=movie['gross'])
plt.title('Budget vs Gross Income')
plt.xlabel('Gross Income')
plt.ylabel('Film Budget')
plt.show()
```

```python
#checking correlation
movie.corr()
```

```python
correlation_matrix = movie.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
movie_numerized = movie
for col_name in movie_numerized.columns:
    if(movie_numerized[col_name].dtype == 'object'):
        movie_numerized[col_name]= movie_numerized[col_name].astype('category')
        movie_numerized[col_name]= movie_numerized[col_name].cat.codes
movie_numerized.head()
```

```python
fig = plt.figure(figsize = (12,6))
correlation_matrix = movie_numerized.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('correlation matrix for numeric features')
plt.xlabel('movie_features')
plt.ylabel('movie_features')
plt.show()
```

```python
movie_numerized.corr()
```

```python
correlation_mat = movie_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_corr_pairs =corr_pairs.sort_values()
sorted_corr_pairs 
```

```python
High_correlated_pairs =sorted_corr_pairs[(sorted_corr_pairs) > 0.5]
High_correlated_pairs
```
